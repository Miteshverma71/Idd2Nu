```python
import os
from PIL import Image
import uuid
import json

folder = "/home/miteshv/Downloads/idd3d/idd3d_seq_10"
```

```python

# ✅ Set your actual paths here
lidar = os.path.join(folder, "lidar")  
source_folder = lidar   
outpath_folder = '/home/miteshv/Downloads/idd3d/idd2nuscenes/sensor/lidar'  # e.g., '/home/user/data/pcd_bin'

# Create outpath folder if it doesn't exist
os.makedirs(outpath_folder, exist_ok=True)

# Check if source folder exists
if not os.path.exists(source_folder):
    raise FileNotFoundError(f"The folder '{source_folder}' does not exist.")

# Process each file in the source folder
for filename in os.listdir(source_folder):
    file_path = os.path.join(source_folder, filename)

    # Convert .pcd or .feather files
    if filename.endswith('.pcd') or filename.endswith('.feather'):
        # Read file in binary mode
        with open(file_path, 'rb') as f:
            binary_data = f.read()

        # Create new .pcd.bin filename
        bin_filename = filename + '.bin'
        bin_file_path = os.path.join(outpath_folder, bin_filename)

        # Write binary data to .pcd.bin file in outpath
        with open(bin_file_path, 'wb') as f:
            f.write(binary_data)

        # Delete original file
        os.remove(file_path)
```

```python
camera = os.path.join(folder, "camera") 

# ✅ Replace these with your actual paths
input_camera_folder = camera
output_camera_folder = '/home/miteshv/Downloads/idd3d/idd2nuscenes/sensor'

# Folder renaming map
folder_mapping = {
    'cam0': 'cam_front_left',
    'cam1': 'cam_back_right',
    'cam2': 'cam_front_right',
    'cam3': 'cam_front',
    'cam4': 'cam_back_left',
    'cam5': 'cam_back'
}

# Create output camera folder
os.makedirs(output_camera_folder, exist_ok=True)

# Process each camera folder
for old_name, new_name in folder_mapping.items():
    old_path = os.path.join(input_camera_folder, old_name)
    new_path = os.path.join(output_camera_folder, new_name)
    os.makedirs(new_path, exist_ok=True)

    # Convert .png to .jpg
    for file in os.listdir(old_path):
        if file.endswith('.png'):
            png_file = os.path.join(old_path, file)
            jpg_file = os.path.join(new_path, os.path.splitext(file)[0] + '.jpg')

            with Image.open(png_file) as img:
                img.convert('RGB').save(jpg_file, 'JPEG')

print("✅ All .png files converted to .jpg and folders renamed in new camera directory.")
```

```python

def generate_sample_json(seq, scene_token, output_path):
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    sample_list = []

    for i in range(100):       #  Change here for multi sequence
        num = f"{i:02d}"
        timestamp = f"{seq}{num}000"
        sample_entry = {
            "token": uuid.uuid4().hex,
            "timestamp": timestamp,
            "prev": "",
            "next": "",
            "scene_token": scene_token
        }
        sample_list.append(sample_entry)

    for i in range(100):      
        if i > 0:
            sample_list[i]["prev"] = sample_list[i - 1]["token"]
        if i < 99:
            sample_list[i]["next"] = sample_list[i + 1]["token"]

    with open(output_path, 'w') as json_file:
        json.dump(sample_list, json_file, indent=4)

    print(f"sample.json created at: {output_path}")
    return sample_list[0]["token"], sample_list[-1]["token"]
```

```python
def generate_scene_json(scene_info, first_token, last_token, output_path):
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    scene_token = uuid.uuid4().hex
    scene_entry = {
        "token": scene_token,
        "log_token": scene_info["log_token"],
        "nbr_samples": 100,
        "first_sample_token": first_token,
        "last_sample_token": last_token,
        "name": scene_info["name"],
        "description": scene_info["description"]
    }

    with open(output_path, 'w') as json_file:
        json.dump([scene_entry], json_file, indent=4)

    print(f"scene.json created at: {output_path}")
    return scene_token
```

```python
import os
import json
import uuid

def generate_sample_annotation_json_from_label_folder(label_folder, output_folder, num_entries=10):
    os.makedirs(output_folder, exist_ok=True)

    # Iterate through each JSON file in the label folder
    for filename in os.listdir(label_folder):
        if filename.endswith('.json'):
            input_path = os.path.join(label_folder, filename)
            output_path = os.path.join(output_folder, f"sample_annotation_{filename}")

            # Load the label JSON file
            with open(input_path, 'r') as f:
                label_data = json.load(f)

            # Extract position, scale, and rotation from the label file
            psr = label_data.get('psr', {})
            position = psr.get('position', {"x": 0, "y": 0, "z": 0})
            scale = psr.get('scale', {"x": 1, "y": 1, "z": 1})
            rotation = psr.get('rotation', {"x": 0, "y": 0, "z": 0})

            # Convert to list format for annotation
            translation = [position['x'], position['y'], position['z']]
            size = [scale['x'], scale['y'], scale['z']]
            rotation_list = [rotation['x'], rotation['y'], rotation['z']]

            annotations = []

            for i in range(num_entries):
                token = uuid.uuid4().hex
                sample_token = uuid.uuid4().hex
                instance_token = uuid.uuid4().hex
                visibility_token = str((i % 5) + 1)
                attribute_tokens = [uuid.uuid4().hex]
                prev = annotations[-1]["token"] if i > 0 else ""
                next = ""

                annotation = {
                    "token": token,
                    "sample_token": sample_token,
                    "instance_token": instance_token,
                    "visibility_token": visibility_token,
                    "attribute_tokens": attribute_tokens,
                    "translation": translation,
                    "size": size,
                    "rotation": rotation_list,
                    "prev": prev,
                    "next": next,
                    "num_lidar_pts": 5 + i,
                    "num_radar_pts": i
                }

                if i > 0:
                    annotations[-1]["next"] = token

                annotations.append(annotation)

            # Save the annotations to the output file
            with open(output_path, 'w') as f:
                json.dump(annotations, f, indent=4)

            print(f"Created: {output_path}")

label_folder = '/path/to/your/label/folder'
output_folder = './annotation'
generate_sample_annotation_json_from_label_folder(label_folder, output_folder)
```
